```python
import matplotlib.pyplot as plt
```

```python
df.plot()
# plt.plot(df.Usage)
```

```python
from statsmodels.tsa.stattools import adfuller
```

```python
result = adfuller(df.Usage)
```

```python
result = adfuller(df.Usage)
if result[1] < 0.05:
    print('It is Stationary')
else:
    print('It is not Stationary')
result[1]
```

```python
result = adfuller(df.Usage.diff().dropna())
if result[1] < 0.05:
    print('It is Stationary')
else:
    print('It is not Stationary')
result[1]
```

```python
df['Diff1'] = df.Usage.diff()
```

```python
df = df.dropna()
df
```

```python
plt.plot(df.Usage,c = 'r',label = 'Original')
plt.plot(df.Diff1,c = 'y',label = 'Difference 1')
plt.legend()
```

# Finding Value Of P

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

plot_pacf(df.Diff1)
```

here lag order is 7 with difference 1

here lag order is 7 with difference  1

# Finding MA

```python
plot_acf(df.Diff1)
```

here MA is 1 With difference 1

```python
from statsmodels.tsa.arima_model import ARIMA

```

```python
# help(ARIMA)

model = ARIMA(df.Usage, order=(7,1,1))#p,d,q
model_fit = model.fit(disp=0)
print(model_fit.summary())
```

```python
residuals = pd.DataFrame(model_fit.resid)
fig, ax = plt.subplots(1,2)
residuals.plot(title="Residuals", ax=ax[0])
residuals.plot(kind='kde', title='Density', ax=ax[1])
plt.show()
```
