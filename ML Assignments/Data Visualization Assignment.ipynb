```python

def scaterplot(x,y,l):
    fig = plt.figure(figsize = (10,5))
    
    plt.scatter(x,y)
#     plt.plot(x,y)
    plt.xlabel(l)
    plt.ylabel('Bodyfat')
#     plt.legend()/
    
```

```python
covid_india = pd .read_csv("..\Datasets\covid_19_india.csv")
covid_india.rename({'State/UnionTerritory':'State'},axis = 1,inplace=True)
print(covid_india.head())
```

```python
# covid_india.value_counts()
x = covid_india.Cured.sum()
y = covid_india.Deaths.sum()
z = covid_india.Confirmed.sum()
state = covid_india.State
x,y,z
```

```python
dir(state)
# help(state)
```

```python
state = covid_india.groupby('State')
```

```python
dir(state)
```

```python
dict = {}

for name in state.groups:
    groupv = state.get_group(name)
    dict[name] = groupv['Confirmed'].sum()
#     print(name,groupv['Cured'].sum())
    
dict 
# for name, group in state:
#     print(name)
# #     dict[name] = group['Cured'].sum() 
#     print(group['Cured'].sum())
```

```python
d = pd.DataFrame({'State':list(dict.keys()),'Confirmed':list(dict.values())})
d1 = d.sort_values(by = ['Confirmed'],ascending = False)
d2 = d1.head(10)
d2['Confirmed']
# list(d2['State'])
```

```python

fig = plt.figure(figsize = (15,5)) 
plt.ylabel("Confirmed")
plt.xlabel('States')
plt.title('Top 10 States of Confirmed Covid 19 ',size=15)

sns.barplot(x= list(d2['State']),y = list(d2['Confirmed']))
```
