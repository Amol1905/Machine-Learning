```python
# df.index
df_month = df.resample("M").mean()
```

```python
df_month
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
plt.xlabel('Vancouver City')
plt.ylabel("Temperature of Vancouver from 2016 to 2018")
ax.bar(df_month['2016':].index, df_month.loc['2016':, "Vancouver"], width=25, align='center')
```

```python
df_month['Vancouver'].plot(figsize = (10,5))
```

```python
week = df.resample("W").mean()
```

```python
week = week.head(1000)
```

```python
fig=plt.figure(figsize=(20,10))
plt.plot(df.loc['2017-01':'2017-08', 'Vancouver'], marker='.', linestyle='-', linewidth = 0.5, label='Daily', color='black')
plt.plot(week.loc['2017-01':'2017-08', 'Vancouver'], marker='o', markersize=8, linestyle='-', label='Weekly', color='coral')
plt.legend()
```

```python
roll_7=df.rolling(7, center=True).mean()
```

```python
fig=plt.figure(figsize=(20,10))
plt.plot(df.loc['2017-01':'2017-08', 'Vancouver'], marker='.', linestyle='-', linewidth = 0.5, label='Daily', color='black')
plt.plot(week.loc['2017-01':'2017-08', 'Vancouver'], marker='o', markersize=8, linestyle='-', label='Weekly', color='r')
plt.plot(roll_7.loc['2017-01':'2017-08', 'Vancouver'], marker='o', markersize=8, linestyle='-', label='Weekly', color='coral')

plt.legend()
```

```python
from pylab import rcParams
import statsmodels.api as sm
rcParams['figure.figsize'] = 11, 9
decomposition = sm.tsa.seasonal_decompose(df_month['Vancouver'], model='Additive')
fig = decomposition.plot()
```

```python
from statsmodels.tsa.stattools import adfuller
result = adfuller(df["Vancouver"])
print('ADF Statistic: %f' % result[0])
print('p-value: %f' % result[1])
print('Critical Values:')
for key, value in result[4].items():
	print('\t%s: %.3f' % (key, value))
```
