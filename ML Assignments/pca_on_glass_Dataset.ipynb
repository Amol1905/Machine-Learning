# PCA On glass Dataset


```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('glass.csv')
```

```python
df.head()
```

```python
x = df
from sklearn.decomposition import PCA
```

```python
model = PCA()
x_scale = model.fit_transform(x)
x_scale
```

```python
plt.figure(figsize=(10,7))

for column in df.columns[1:9]:
    plt.scatter(df[column],df['RI'],label = str(column))
    plt.xlabel("RI")
    plt.legend()
```

# After PCA Data is

```python

plt.figure(figsize=(10,7))

for column in range(1,x_scale.shape[1]):
    plt.scatter(x_scale[0],x_scale[column],label = str(column))
    plt.xlabel("RI")
    plt.legend()
```
