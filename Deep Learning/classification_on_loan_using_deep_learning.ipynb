```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# df = pd.read_csv('/content/diabetes.csv')
df = pd.read_csv('/content/loan_data_set.csv')
```

```python
df.head()
```

```python
x = df.drop(columns=['Loan_Status','Loan_ID'],axis  = 1)
y = df['Loan_Status']
```

```python
from keras.models import Sequential
from keras.layers import Dense
model = Sequential()
# $\ #  Layers

model.add(Dense(12,activation  = 'relu',input_dim = 11))
model.add(Dense(10,activation  = 'relu'))
model.add(Dense(1,activation  = 'sigmoid'))# Output layer
model.summary()
```

```python
# binary entropy,hinge loss,squared loss
model.compile(loss = 'binary_crossentropy',optimizer = 'adam',metrics=['accuracy'])
# model.compile(loss= )
```

```python
# epochs = 50
# batch =10
history = model.fit(x,y,batch_size=10,epochs=50)
```

```python
loss,acc  = model.evaluate(x,y)
```
