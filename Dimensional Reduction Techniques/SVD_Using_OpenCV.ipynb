```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('img.jpg')
img
```

```python
plt.imshow(img)
```

```python
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.imshow(gray,cmap = 'gray')
```

```python
U,S,V = np.linalg.svd(gray,full_matrices  = False)
```

```python
comps = [U.shape[0],100,50,30,20,10,5,2]
plt.figure(figsize = (15,10))
for i in range(len(comps)):
#     print(i)
    trans_image = U[:,:comps[i]]@(np.diag(S[:comps[i]]))@(V[:comps[i],:]) 
    plt.subplot(2,4,(i+1))
    plt.imshow(trans_image,cmap = 'gray')
    plt.title(f'image with n_components {comps[i]}')
#     trans_image
    
```
