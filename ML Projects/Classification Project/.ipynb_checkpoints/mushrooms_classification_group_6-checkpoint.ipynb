```python
def LabelEncoder(data):
    for column in df1.columns:
        steps = 1 / (len(data[column].unique())-1)
        print(f'{column} :{steps}')
        value = 0
        val = format(0.0000,'4f')
        for i in data[column].unique():
            for letter in range(len(data[column])):
                if data[column][letter] == i:
                    df[column][letter] = val
            value += steps   
            val = format(value,'4f')
```

# 4. Preparing Data for Machine Learning

```python
x = df.drop(columns = ['class']).values
y = df['class'].values
x.shape,y.shape
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.2)
```

```python
model1 = []
accuracy = []
auc = []
def Model(model):
    model.fit(xtrain,ytrain)
    pred = model.predict(xtest)
    acc  = accuracy_score(ytest,pred)
    print(f'Accuracy: {acc}')
    print(classification_report(ytest,pred))
    model1.append(str(model))
    accuracy.append(acc)
    fig, axes = plt.subplots(1, 2, figsize=(15, 5))
    fig.suptitle(f'{model} Plots')
    #probability of positive class
    pred_prob=model.predict_proba(xtest)[:,1]
    fpr,tpr,_=metrics.roc_curve(ytest,pred_prob)
    auc1=metrics.roc_auc_score(ytest,pred_prob)
    auc.append(auc1)
    
#     plt.plot(fpr,tpr,label="ROC curve")
    sns.heatmap(confusion_matrix(ytest,pred),annot = True,ax = axes[0])
    axes[0].set_title('Confusion Matrix')
    sns.lineplot(fpr,tpr,ax = axes[1])
#     axes[1].set_title('ROC Curve')
    plt.plot(fpr,tpr,label="ROC curve")
    plt.title('ROC Curve')
    plt.xlabel("FPR")
    plt.ylabel("TPR")
    plt.legend() 
    return pd.DataFrame({'Model':model1,'Accuracy':accuracy,'AUC':auc})


```

```python
model = LogisticRegression()
result = Model(model)
# model
```

# 2. KNN 

### Finding Value fo K for KNN


```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import RepeatedStratifiedKFold,GridSearchCV

```

```python
#parameter grid
#n_neighbors,metric,weights
n_neighbors=range(1,31)
metric=["euclidean","manhattan","minkowski"]
weights=['uniform','distance']
grid={"n_neighbors":n_neighbors,"metric":metric,"weights":weights}
```

```python
def Best_param(model,grid):
    kfold=RepeatedStratifiedKFold(n_splits=5,n_repeats=3,random_state=1)
    grid_cv=GridSearchCV(estimator=model,param_grid=grid,cv=kfold,scoring="accuracy",n_jobs=1)
    #result
    res=grid_cv.fit(xtrain,ytrain)
    return res.best_params_,res.best_score_

```

```python
Best_param(KNeighborsClassifier(),grid)
```

```python
model_best=KNeighborsClassifier(metric='manhattan',n_neighbors=1,weights='uniform')
result = Model(model_best)
```
